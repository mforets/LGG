## The 5-dimensional case

````python
n = 5 

A = random_matrix(RDF, n)

# time horizon 
T = 2

# initial condition 
X0 = BoxInfty(center = [1,0,0,0,0], radius = 0.01)

# time discretization 
tau = 1e-1
````

````python
%%time
fp = compute_flowpipe(A, X0, time_horizon = T, time_step = tau, solver='Gurobi')
````

For $n=5$, it collapses. We get the error: 

```
GLPKError: glp_set_rii: i = 79; rii = 0; invalid scale factor
```

Let us try using Gurobi optimization. We also get into trouble: Gurobi gives the error: 

```
MIPSolverException: Gurobi: The problem is infeasible or unbounded
```

It is indicated that the problem is from `_Omega_i_supports(d)`, in the line `supp_fun_polyhedron(Omega0, d, solver=solver, verbose=verbose)`. We have the option of using the `verbose` keyword to display the optimization problem.

````python
%%time
fp = compute_flowpipe(A, X0, time_horizon = T, time_step = tau, solver='Gurobi', verbose=True)
````

We see that it want to find the minimum of `-x0`, and the constraints are huge, some can be $10^{180}$. This is probably the source of numerical instability, and the question is why does it generate such big numbers.

Since `len(P12.V)` gives 1024, it means that it has computed Minkowski addition by taking the sums over all possible vertices. We know that this is correct but it is not optimal, since it includes (possibly many) redundant vertices. I don't  know if there is a command to remove the redundant vertices in MPT (not in [this doc](http://control.ee.ethz.ch/~mpt/3/Geometry/OperationsWithPolyhedra) at least).

What comes to mind is to use this simple implementation in Sage, and then try to build a Polyhedron from those vertices.

We need to take sums of vertices, and be careful because the sum `+` symbol executes concatenation over lists. A simple workaround is to translate to Sage vectors.

````python
v1[0]
````

````python
vector(v1[0]) + vector(v1[0])
````

````python
list(vector(v1[0]) + vector(v1[0]))
````

We now construct the set of all possible sums of vertices of each polytope. 

````python
v1v2 = []
for u in v1:
    for v in v2:
        v1v2.append(list(vector(u) + vector(v)))
````

````python
v1v2[0]
````

````python
len(v1v2)
````

When we do `Polyhedron(vertices = v1v2)`, we get the same problem: `ValueError: *Error: Possibly an LP cycling occurs.  Use the Criss-Cross method.`. Maybe that is because by default it tries to compute *both* the H-representation and the V-representation, and it fails to compute the H-representation for the default algorithm (Dual Simplex) in this case.

- [Criss-cross algorithm](https://en.wikipedia.org/wiki/Criss-cross_algorithm)

It is strange, because in Matlab it does compute the H-representation. 

Let us try in Julia.

````python
from scipy.io import savemat

d=dict()

d['v1'] = v1; 
d['v2'] = v2;
d['v1v2'] = v1v2;

savemat('experimental_polytope_vertices_2.mat', d)
````
