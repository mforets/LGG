```python
from LGG.lgg import compute_flowpipe, plot_flowpipe
```

```python

A = matrix([[1., -2.],[0.3, 1.]])

B = identity_matrix(2)

# time horizon 
T = 3

# input domain U
U = Polyhedron(vertices = [[-.1,-.1],[.1,.1],[-.1,.1],[.1,-.1]])

# initial condition 
X0 = Polyhedron(vertices = [[0.5,0.5],[0.5,1.],[0.75,1.5],[1.,1.],[1.,0.5]])

# time discretization 
tau = 3/100;
```

```python
%%time
fp = compute_flowpipe(A, X0, B, U, time_horizon = T, time_step = tau, directions={'select':'random', 'order':50}, 
                      solver='GLPK', base_ring=QQ)
```

```python
# plot the result
plot_flowpipe(fp)
```

```python
plotFP = Graphics()
for p in fp:
    #plotFP += p.plot(point=False, line='black', fill=hue(0.6))
    plotFP += p.plot(point=False, fill=False, line='black')
```
